```python
import numpy as np
import glob
import h5py
import cv2
from keras.preprocessing.image import load_img, img_to_array


def make (target_size = (64,64),gray = False):
    
    inpath = "./input"
    outpath = "./output"


    orgs = []
    masks = []

    print("mask img")
    files = glob.glob(inpath + "/mask/*.jpg")
    for imgfile in files:
        print(imgfile)
        img = load_img(imgfile, target_size = target_size)
        imgarray = img_to_array(img)
        masks.append(imgarray)

    if(gray):
        print("gray img")
        for fname in files:
            #fname = fname[]
            I = cv2.imread(fname)
            kernel = np.ones((5,5), np.uint8)
            dilation = cv2.dilate(I, kernel, iterations = 1)
            diff = cv2.subtract(dilation, I)
            diff_inv = 255 -diff
            diff_inv_binarized = cv2.cvtColor(diff_inv, cv2.COLOR_BGR2GRAY)
            cv2.imwrite(inpath +"/orignal/"+fname[13:], diff_inv_binarized)


    print("orignal img")
    files = glob.glob(inpath + "/orignal/*.jpg")
    for imgfile in files:
        print(imgfile)
        img = load_img(imgfile, target_size = target_size)
        imgarray = img_to_array(img)
        orgs.append(imgarray)

    perm = np.random.permutation(len(orgs))
    orgs = np.array(orgs)[perm]
    masks = np.array(masks)[perm]
    threshold = len(orgs)//10*9
    imgs = orgs[:threshold]
    gimgs = masks[:threshold]
    vimgs = orgs[threshold:]
    vgimgs = masks[threshold:]
    print('shapes')
    print('org imgs  : ', imgs.shape)
    print('mask imgs : ', gimgs.shape)
    print('test org  : ', vimgs.shape)
    print('test tset : ', vgimgs.shape)

    outh5 = h5py.File(outpath+'/datasetimages.hdf5', 'w')
    outh5.create_dataset('train_data_raw', data=imgs)
    outh5.create_dataset('train_data_gen', data=gimgs)
    outh5.create_dataset('val_data_raw', data=vimgs)
    outh5.create_dataset('val_data_gen', data=vgimgs)
    outh5.flush()
    outh5.close()
```

```python
make(target_size = (128,128))
```
